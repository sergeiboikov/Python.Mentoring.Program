## Используя данные из файла olympics.csv подготовьте jupyter notebook со следующими выборками и графиками: 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df = pd.read_csv('olympics.csv', skiprows=4)
```

- Какая страна выйграла большинство золотых медалей мужчинами в бадминтоне? Отсортировать результат в алфавитном порядке по имени игрока.

```python
#Prepare filtered dataframe
filtered_df = df[(df["Discipline"] == "Badminton") & (df["Gender"] == "Men") & (df["Medal"] == "Gold")]

#Count medals for each NOC from filtered_df
noc_medals = filtered_df \
    .groupby("NOC").size() \
    .reset_index(name="noc_medals") \
    .sort_values(["noc_medals"],
                 ascending=False)

#Find top 1 NOC value
noc_top_1 = noc_medals.head(1)[["NOC"]].values[0][0]

#Prepare final dataframe 
result_df = filtered_df[(filtered_df["NOC"] == noc_top_1)][["NOC", "Athlete"]] \
    .sort_values(["Athlete"])

#Output
result_df
```

- Какие три страны выйграли большинство медалей в последние годы (с 1984 по 2008)?

```python
#Prepare filtered dataframe
filtered_df = df[(df["Edition"] >= 1984) & (df["Edition"] <= 2008)]

#Count medals for each NOC from filtered_df
noc_medals = filtered_df \
    .groupby("NOC").size() \
    .reset_index(name="noc_medals") \
    .sort_values(["noc_medals"], ascending=False)

#Find top 3 NOC value
noc_medals.head(3)[["NOC", "noc_medals"]]

```

- Используя groupby(), постройте график число всех медалей выйгранных на каждой олимпиаде.

```python
# Prepare dataframe
medals_total = df \
    .groupby("Edition").size() \
    .reset_index(name="medals_total")
x = medals_total["Edition"]
y = medals_total["medals_total"]

# Define plot
plt.plot(x, y, label='Total Medals')
plt.show()
```

- Создайте список показывающий число всех медалей выйгранных каждой страной в течение всей истории олимпийских игр. Для каждой страны, необходимо показать год первой и последней заработанной медали.

```python
# Dataframe with first editions for each NOC
noc_first_edition = \
    df.assign(rn=df.sort_values(["Edition"]) \
              .groupby(["NOC"]).cumcount() + 1, \
              first_edition=df["Edition"]).query("rn == 1") \
[["NOC", "first_edition"]]

# Dataframe with last editions for each NOC
noc_last_edition = \
    df.assign(rn=df.sort_values(["Edition"], ascending=False) \
              .groupby(["NOC"]).cumcount() + 1, \
              last_edition=df["Edition"]).query("rn == 1") \
[["NOC", "last_edition"]]

# Dataframe with count of medals for each NOC
noc_medals_total = df \
    .groupby("NOC").size() \
    .reset_index(name="medals_total")

# Join results
final_df = pd \
    .merge(pd \
           .merge(noc_first_edition, noc_last_edition, on="NOC"), \
           noc_medals_total, on="NOC") \
    .sort_values("NOC")

# Output
final_df
```

- Постройте график числа золотых медалей выйгранных США мужчинами и женщинами в атлетике?

```python
gold_men_df = df[
    (df["Discipline"] == "Athletics") & (df["Gender"] == "Men") \
    & (df["NOC"] == "USA") & (df["Medal"] == "Gold")] \
    .groupby("Edition").size() \
    .reset_index(name="count_medals")

gold_women_df = df[
    (df["Discipline"] == "Athletics") & (df["Gender"] == "Women") \
    & (df["NOC"] == "USA") & (df["Medal"] == "Gold")] \
    .groupby("Edition").size().reset_index(name="count_medals")

# Define subplots
fig, ax1 = plt.subplots()
ax1.set_xlabel('Edition')
ax1.set_ylabel('Count medals')

# Men
color = 'tab:red'
x = gold_men_df["Edition"]
y = gold_men_df["count_medals"]
ax1.plot(x, y, color=color)

# Women
color = 'tab:blue'
x = gold_women_df["Edition"]
y = gold_women_df["count_medals"]
ax1.plot(x, y, color=color)

# Add legend
plt.legend(["Men", "Women"], loc="upper right")

# Output
plt.show()
```

- Постройте график 5 атлетов которые выйграли большинство золотых медалей.

```python
top_5_athletes_df = df[(df["Medal"] == "Gold")] \
    .groupby("Athlete").size() \
    .reset_index(name="medals_total") \
    .sort_values("medals_total", ascending=False) \
    .head(5)

#Output
ax = top_5_athletes_df.plot.bar(x="Athlete", y="medals_total")
```

- Постройте таблицу в которой по годам всех олимпиад покажите топовых атлетов США(1 атлет на год) по общему количеству медалей? Включите дисциплину атлета.

```python
filtered_df = df[(df["NOC"] == "USA")]
top_athlets_usa_df = df.assign(rn=filtered_df \
                               .groupby(["Edition"]) \
                               .cumcount() + 1).query("rn == 1") \
[["Edition", "Discipline", "Athlete"]]

top_athlets_usa_df
```
